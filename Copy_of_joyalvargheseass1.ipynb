```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
### Reading Dataset
```

```python
data = pd.read_excel('iris.xls')
```

```python
data.head()
```

```python
### Reading Columns

data.columns
```

```python
### Mean of each columns

SL_mean_data = data["SL"].mean()
SW_mean_data = data["SW"].mean()
PL_mean_data = data["PL"].mean()
PW_mean_data = data["PW"].mean()

print("SL mean:",SL_mean_data, "\nSW Mean:", SW_mean_data, "\nPL mean:",PL_mean_data, "\nPL mean:",PL_mean_data)
```

```python
### Checking for null values
data.isnull()
```

```python
data.isnull().sum()
```

```python
### visuallisations
```

```python
plt.figure(figsize=(8,6))


plt.subplot(2,1,1)
plt.scatter (data['SL'], data['PL'], s=18, c='red')
plt.title('SL vs PL')
plt.xlabel('SL')
plt.ylabel('PL')


plt.subplot(2,1,2)
plt.scatter (data['SW'], data[ 'PW'], s=20, c='blue')
plt.title('SW vs PW')
plt.xlabel('SW')
plt.ylabel('PW')
plt.tight_layout()
```
